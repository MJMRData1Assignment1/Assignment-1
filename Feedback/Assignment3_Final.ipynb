## Exercise 4: Shape the data

- **(Stretch Goal)** Examine how many books were published at each year and find lowest, highest rate of each year. 

```python
# your code here
# use many cells if you like to structure your code well

# define a function to apply to each cell later
def get_author(url):

    # split the url string by '/' then '.' 
    name = url.split('/')[-1].split('.')[1:] 

    return name

```

```python
# Apply the function to 'author_url' column using '.map'
goodreads_new['author'] = goodreads_new.author_url.map(get_author)
goodreads_new['author']
```

```python
# check the data type of each column, make sure they are comparable
goodreads_new.dtypes
```

## Exercise 6: Investigate the relationship between the number of reviews and the average rating for books in the dataset cleaned-goodreads.csv procided.

- Calculate the correlation coefficient. Give me a short definition of this coefficient
- Create a scatter plot showing the relationship between these two features.
- Based on the plot and the correlation, provide a brief interpretation of the relationship.

### Python Tools: Use pandas and numpy for correlation, and matplotlib or seaborn for the scatter plot.

```python
# your code here
# use many cells if you like to structure your code well

# load the csv file procided
cleaded_goodreads = pd.read_csv(r"C:\Users\Admin\Desktop\Code\lab-02\data\cleaned-goodreads.csv")
cleaded_goodreads
```

```python
# sort the data by rating
cleaded_goodreads = cleaded_goodreads.sort_values(by='rating', ascending=True)
cleaded_goodreads
```

```python
x_mean = cleaded_goodreads.rating.mean()
y_mean = cleaded_goodreads.review_count.mean()
print(x_mean, y_mean)
```

```python
# calculate xi - x_mean and yi - y_mean
def numeratior_x(xi):
    return xi - x_mean
cleaded_goodreads['x1'] = cleaded_goodreads.rating.map(numeratior_x)

def numeratior_y(yi):
    return yi - y_mean
cleaded_goodreads['y1'] = cleaded_goodreads.review_count.map(numeratior_y)

# calculate product of two columns
cleaded_goodreads['x1y1'] = cleaded_goodreads['x1'] * cleaded_goodreads['y1']

# difference of two squares
def denominator_x(xi):
    return (xi - x_mean) **2
cleaded_goodreads['x2'] = cleaded_goodreads.rating.map(denominator_x)

def denominator_y(yi):
    return (yi - y_mean) **2
cleaded_goodreads['y2'] = cleaded_goodreads.review_count.map(denominator_y)

# Calculate the correlation coefficient
numerator = np.sum(cleaded_goodreads.x1y1)
denominator = np.sqrt(np.sum(cleaded_goodreads.x2)) * np.sqrt(np.sum(cleaded_goodreads.y2))
correlation_coefficient = numerator/denominator

print(correlation_coefficient) 
```

```python
# this coefficient indicates nearly no linear relationship
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# create the scatter plot 
sns.scatterplot(data=cleaded_goodreads, x='rating', y='review_count')
plt.show()
```

```python
# Brief interpretation of the relationship between the number of reviews and the average rating of books
```

Considering the scatter plot, one can observe that with a higher number of reviews the average rating of the respective book becomes closer to the mean, with a significant number of reviews (y>10,000) leading to a dispersion only between a rating of 3.5 and 4.5 except from a handful of outliers. One can interpretate this as frequently reviewed books being more weightedly rated and considering a normal distribution of values the average will reach closer to the mean. Another interpretation could be that frequently reviewed books in general score quite well (average rating around 4), where the quality of the book makes it that it is being read and reviewed so often.
