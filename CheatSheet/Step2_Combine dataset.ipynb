Combine the data from .csv to GeoDataFrame

```python
# import the library

import pandas as pd
import numpy as np
```

```python
# assign the data source file path
file_path = 'C:\\Users\\Admin\\Desktop\\Code\\source_file.csv'

# load the data source file(.csv)
data_source = pd.read_csv(file_path)
data_source.head()
```

```python
# select the data want to be combined
data_select = data_source[['LAD14CD','Electorate']]
data_select.head()
```

```python
data_select['new_data'] = data_select['Electorate']
data_select = data_select.drop(columns=['Electorate'])
data_select.head() # prepare the dataset will be combined
```

```python
import geopandas as gpd

# load the main file(.shp
file_path = 'C:\\Users\\Admin\\Desktop\\Code\\brexit.gpkg'
main_dataset = gpd.read_file(file_path)
main_dataset.head()
```

```python
# merge the dataset based on common column 'LAD14CD'

merged_gdf = main_dataset.merge(data_select[['LAD14CD', 'new_data']], on='LAD14CD', how='left')

"""
main_dataset is (.shp)file
data_select is (.csv)file
on = 'common column'
"""

merged_gdf.head()
```

Transform DataFram(.csv) to GeoDataFrame(.shp)

```python
import geopandas as gpd
from shapely import wkt

# transform WKT column to geometry
data_source['geometry'] = data_source['geometry_wkt'].apply(wkt.loads)

# build GeoDataFrame
gdf = gpd.GeoDataFrame(data_source, geometry='geometry')
gdf.head()
# set coordinate
# gdf.set_crs(epsg=4326, inplace=True)
```

Transform GeoDataFrame (.shp) to DataFrame (.csv)

```python
import geopandas as gpd

file_path = 'C:\\Users\\Admin\\Desktop\\Code\\brexit.gpkg'
dataset = gpd.read_file(file_path).set_index('LAD14CD', drop=False)
dataset.head()
```

```python
# select target columns
source_file = dataset[['LAD14CD','Area','Electorate','geometry']]
source_file.head()
```

```python
# save geometry with GeoDataFrame to WKT (Well-Known Text)
source_file['geometry_wkt'] = source_file.geometry.apply(lambda geom: geom.wkt)
source_file.head()
```

```python
# Drop unnecessary columns
source_file = source_file.drop(columns=['geometry','LAD14CD'])
source_file.head()
```

```python
# reset the index
source_file.reset_index(inplace=True)
source_file.head()
```

```python
# save the csv file
source_file.to_csv('C:\\Users\\Admin\\Desktop\\Code\\source_file.csv', index=False)
```
