```python
Main_dataset = pd.read_csv(folder_dir+'\\Geometry of Wijks.csv', encoding='utf-8')
Main_dataset.head()
```

```python
Main_dataset['CBS_Wijkcode'] = Main_dataset['CBS_Wijkcode'].str.strip('"')
Main_dataset['Wijkcode'] = Main_dataset['Wijkcode'].str.strip('"')
Main_dataset['Wijk'] = Main_dataset['Wijk'].str.strip('"')
Main_dataset['Gebiedcode'] = Main_dataset['Gebiedcode'].str.strip('"')
Main_dataset['Gebied'] = Main_dataset['Gebied'].str.strip('"')
Main_dataset['Stadsdeelcode'] = Main_dataset['Stadsdeelcode'].str.strip('"')
Main_dataset['Stadsdeel'] = Main_dataset['Stadsdeel'].str.strip('"')

# change the header name or dtype
Main_dataset.rename(columns={'Wijkcode': 'Area code', 'Wijk': 'Area'}, inplace=True)

Main_dataset.head()
```

```python
# merge the dataset based on common column 'Area code'
merged_gdf = Main_dataset.merge(Rent_Average_Source[['Area code', 'Rent: average','Average house-value for tax purposes']], on='Area code', how='left')
merged_gdf.head()
```

```python
merged_gdf = merged_gdf.merge(Public_Spaces_Source[['Area code', 'Area public green (ha)']], on='Area code', how='left')
merged_gdf = merged_gdf.merge(Traffic_Source[['Area code', 'Parking spots total']], on='Area code', how='left')
merged_gdf = merged_gdf.merge(Income_Source[['Area code', 'Mean disposable household income']], on='Area code', how='left')
merged_gdf = merged_gdf.merge(Socio_economic_status[['Area code', 'Most vulnerable (%)']], on='Area code', how='left')
merged_gdf = merged_gdf.merge(Safety_Source[['Area code', 'Safety index: High Impact Crime']], on='Area code', how='left')

merged_gdf.head()
```

```python
merged = merged_gdf[['OBJECTNUMMER','CBS_Wijkcode','Area code','Area','Rent: average','Average house-value for tax purposes','Area public green (ha)','Parking spots total','Mean disposable household income','Most vulnerable (%)','Safety index: High Impact Crime','geometry','LNG','LAT','Gebiedcode','Gebied','Stadsdeelcode','Stadsdeel','Oppervlakte_m2']]
merged.head()
```

```python
merged.info()
```

```python
merged['Rent: average'] = merged['Rent: average'].apply(pd.to_numeric, errors='coerce')
merged['Average house-value for tax purposes'] = merged['Average house-value for tax purposes'].apply(pd.to_numeric, errors='coerce')
merged['Area public green (ha)'] = merged['Area public green (ha)'].apply(pd.to_numeric, errors='coerce')
merged['Average house-value for tax purposes'] = merged['Average house-value for tax purposes'].apply(pd.to_numeric, errors='coerce')
merged['Mean disposable household income'] = merged['Mean disposable household income'].apply(pd.to_numeric, errors='coerce')
merged['Safety index: High Impact Crime'] = merged['Safety index: High Impact Crime'].apply(pd.to_numeric, errors='coerce')

merged.info()
```

```python
# save the dataset
merged.to_csv(folder_dir+'\\MergedData.csv') 
```
