```python
# Getting the author_url
test_string = df.author_url[:]
test_string
```

```python
# a function to get URL and give us Authors name

def get_author(url):

    author = url.split('/')[-1].split('.')[1:][0]
     
    return author

```

```python
# Using '.map' to apply the get_author function to the 'author_url' column. Create a new column 'author' to contain the names.
df['author'] = df.author_url.map(get_author)

# to test if the code is working
df.author[0:5]
```
